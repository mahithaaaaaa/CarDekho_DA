```python
import pandas as pd 
df = pd.read_csv("car_Dekho_DA.csv")
df.head()
```

```python
df = pd.DataFrame(df)
```

```python
max_year = df["Year"].max()
min_year = df["Year"].min()
print(f"the years ranges from {min_year} to {max_year}") 
```

```python
highest_price_vehicle_sold = df["Selling_Price"].max()
print(f"The highest price a vehicle sold is {highest_price_vehicle_sold} lakhs")
```

```python
import matplotlib.pyplot as plt
df['Fuel_Type_Code'] = df['Fuel_Type'].astype('category').cat.codes
plt.scatter(df.Kms_Driven, df.depriciation, c=df['Fuel_Type_Code'], cmap='viridis', alpha=0.7)
plt.xlabel("Kilometers Driven")
plt.ylabel("Cost Depriciation in Lakhs")
plt.title("Cost Depriciation vs Kilometers Driven")
cbar = plt.colorbar()
cbar.set_label('Fuel Type')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Kms_Driven, df.Selling_Price, c='blue', alpha=0.5)
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling Price in Lakhs")
plt.title("Selling Price vs Kilometers Driven")
plt.show()
```

```python
two_wheelers = df[df["Present_Price"]<3.5]
print(two_wheelers.head())
```

```python
two_wheelers.reset_index(drop=True,inplace=True)
two_wheelers.head()
```

```python
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "alto k10"]
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "800"]
two_wheelers = two_wheelers.loc[two_wheelers.Car_Name != "omni"]
two_wheelers.head()
```

```python
two_wheelers.reset_index(drop=True,inplace=True)
```

```python
two_wheelers.head()
```

```python
two_wheelers.loc[two_wheelers.Year == two_wheelers.Year.min()]
```

```python
two_wheelers.loc[two_wheelers.Year == two_wheelers.Year.max()]
```

```python
import matplotlib.pyplot as plt
box = plt.boxplot(two_wheelers.Selling_Price, patch_artist=True)
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.ylabel("Selling Price in Lakhs")
plt.title("Boxplot of Selling Price for Two Wheelers")
plt.show()
```

```python
two_wheelers[two_wheelers.Selling_Price>1.6]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig_dims = (12, 20)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(y='Car_Name', data=two_wheelers, ax=ax, palette='viridis')
ax.set_xlabel('Count')
ax.set_ylabel('Car Name')
ax.set_title('Count of Car Names')
plt.show()
```
